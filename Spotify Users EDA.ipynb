# Data Visualization

```python
df['Age'].value_counts()
```

```python
df['Age'] = df['Age'] + ' years'
```

```python
age_distribution = df['Age'].value_counts()
plt.figure(figsize = (8,8))
plt.title('Age Distribution', size = 25 )
plt.pie(age_distribution, autopct = '%.2f', labels = age_distribution.index)
plt.show();
```

### Most of the users of Spotify are adults and teenagers where about 4/5 th part occupied by adults ranging from age 20-35 years.

```python
df['Gender'].value_counts()
```

```python
gender_distribution = df['Gender'].value_counts()
plt.figure(figsize = (8,8))
plt.title('Gender Distribution', size = 25 )
plt.pie(gender_distribution, autopct = '%.2f', labels = gender_distribution.index, colors = sns.color_palette('Set2'))
plt.show();
```

### 3/4 th of the Spotify users are Females.

## Summary

1. The majority of Spotify users are adults and teenagers, with adults aged 20-35 making up the largest portion.
2. Approximately 75% of Spotify users are female.
3. Most Spotify users have been consistent users for more than 6 months, with a significant portion using the platform for over 2 years. Female users are the most prominent in all time periods.
4. Spotify users primarily use smartphones for listening, followed by computers/laptops, smart speakers, or voice assistants.
5. The majority of Spotify users prefer the free (ad-supported) plan, but some are willing to switch to the premium plan.
6. Among users interested in purchasing the premium plan, the individual plan is the most preferred, followed by the duo plan.
7. Over 75% of users primarily use Spotify for listening to music rather than podcasts.
8. The most preferred music genres by Spotify users are melody, classical, pop, and rap.
9. Around 60% of users prefer listening to music at night, followed by the afternoon and morning.
10. Spotify users listen to music for relaxation, stress relief, uplifting, and motivation.
11. Users listen to music while traveling, during leisure time, workouts, study hours, and office hours.
12. Users primarily discover music through recommendations, playlists, and radio.
13. Spotify music recommendations are generally well-received, with ratings mostly between 3 to 5.
14. Podcasts are not listened to on a daily basis by most users.
15. The most listened-to podcast genres by Spotify users are comedy, lifestyle and health, health and fitness, sports, and food and cooking.
16. Users prefer storytelling podcasts, followed by conversational, interview, and educational formats.
17. Users enjoy podcasts from both known individuals and unknown podcasters.
18. Shorter podcasts are often preferred by users.
19. Overall, Spotify users are satisfied with the available and recommended podcasts on the platform.
